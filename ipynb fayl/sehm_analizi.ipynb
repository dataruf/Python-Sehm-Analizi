```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
sehmler = ["AAPL", "MSFT"]
```

```python
data = yf.download(sehmler, period="1y")
baglanis = data["Close"]
plt.figure(figsize=(14, 7))
for sehm in sehmler:
    plt.plot(baglanis[sehm], label= f"{sehm} Baglan")
plt.title("AAPL & MSFT Shm Analizi (balan qiymtlti sasnda)", fontsize = 14, weight = "bold")
plt.xlabel("Tarix")
plt.ylabel("Qiymt (USD)")
plt.legend()
plt.grid(True)
plt.savefig("sehmanaliz.png")
plt.show()
```
